````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

%matplotlib inline
````

````python
sns.pointplot(data=previsao_renda, x='posse_de_imovel', y='renda', dodge=True, ci=90)
````
