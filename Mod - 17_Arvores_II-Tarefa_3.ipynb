### 1. Carregar as bases

Vamos utilizar nesta tarefa as bases de reconhecimento de atividade humana atravs do celular. Carregue novamente as bases salvas na tarefa I.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, roc_auc_score
from sklearn.model_selection import GridSearchCV
````

````python
# train
X_train = pd.read_csv('./input/X_train.csv', index_col=[0, 1])
y_train = pd.read_csv('./input/y_train.csv', index_col=[0, 1])

# test
y_test = pd.read_csv('./input/y_test.csv', index_col=[0, 1])
X_test = pd.read_csv('./input/X_test.csv', index_col=[0, 1])
````

### 2. Calcule os ```ccp_alphas```. 

Vamos seguir uma lgica bem em linha com o que j estamso acostumados, com as seguintes orientaes:

- Utilizar treinamento e teste conforme j vieram definidos originalmente
- Por pragmatismo, utilizar ```min_samples_leaf=20```
- Utilize as mesmas "3 melhores variveis" identificadas no exerccio anterior.

````python
X_train_cut = X_train[['57 tGravityAcc-energy()-X', '272 fBodyAcc-mad()-X', '560 angle(Y,gravityMean)']]
X_test_cut = X_test[['57 tGravityAcc-energy()-X', '272 fBodyAcc-mad()-X', '560 angle(Y,gravityMean)']]
````

````python
%%time
path = DecisionTreeClassifier(random_state=2360873, min_samples_leaf=20).cost_complexity_pruning_path(X_train_cut, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
ccp_alphas = np.unique(ccp_alphas[ccp_alphas>=0])
ccp_alphas
````

### 3. *Grid Search*

Vou deixar especificaes iniciais mnimas, que visam limitar o tempo de mquina, pois um procedimento desses pode demorar muito tempo dependendo da especificao. Mas conforme voc for ficando confortvel com o tempo consumido pelo procedimento, pode fazer um algoritmo mais exaustivo, por exemplo, avaliando mais valores de ```ccp_alpha```.

- Mea o tempo
- Utilize a base de treinamento
- Utilize um *cross validation* do tipo *k-fold*, especifique k=10
- Voc pode ler 1 a cada ```k``` valores para uma melhor varredura utilizando, por exemplo, ```ccp_alpha[::10]```
- No se esquea de limitar o nmero de variveis

````python
%%time
clf = DecisionTreeClassifier(random_state=2360873)
# grid_parametros = {'ccp_alpha': ccp_alphas[::5]}
# grid_parametros = {'ccp_alpha': ccp_alphas[::10]}
grid_parametros = {'ccp_alpha': ccp_alphas}
grid_parametros
````

````python
%%time
grid = GridSearchCV(estimator=clf, param_grid=grid_parametros, cv=10, verbose=100)
grid.fit(X_train_cut, y_train)
````

````python
resultados = pd.DataFrame(grid.cv_results_)
resultados.head()
````

````python
print(grid.best_index_)
print(grid.best_params_)
print(grid.best_score_)
````

### 4. Avaliando a rvore

- Obtenha a rvore na melhor configurao treinada em toda a base de treino
- Calcule a acurcia dessa rvore na base de testes
- Visualize a matriz de confuso

````python
melhor_ccp = resultados.iloc[grid.best_index_, 4]
clf = DecisionTreeClassifier(random_state=2360873, ccp_alpha=melhor_ccp).fit(X_train_cut, y_train)
clf.score(X_test_cut, y_test)
````

````python
ConfusionMatrixDisplay.from_estimator(clf, X_test_cut, y_test)
````

### 5. Melhorando a rvore

- Observe a importncia das variveis, e selecione as 3 com maior importncia
