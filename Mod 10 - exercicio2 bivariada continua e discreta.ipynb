````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
previsao_renda = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
previsao_renda.info()
````

````python
previsao_renda1 = previsao_renda[['posse_de_imovel','posse_de_veiculo','renda']]
previsao_renda1.pivot_table(index=['posse_de_imovel','posse_de_veiculo'], 
                    values='renda', 
                    aggfunc='mean', 
                    margins=True)
````

````python
sns.pointplot(x="posse_de_imovel", y="renda",
              data=previsao_renda1, dodge=True, ci = 100)
plt.show()
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=previsao_renda1, dodge=True, ci = 100)
plt.show()
````

````python
sns.lmplot('posse_de_imovel', 'posse_de_veiculo', hue = 'renda', data = previsao_renda1, fit_reg = False)
````
